```python
import pandas as pd
```

```python
import sqlite3

# create a new database connection
conn = sqlite3.connect('data/zips.db')

# close the connection
conn.close()
```

```python
df = pd.read_excel('data/census_data_ZIPcodes.xlsx')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['ZIP'] = df['ZIP'].astype( str )
```

```python
#add leading zeros
for i in range(len(df['ZIP'])):
    df['ZIP'][i] = df['ZIP'][i].zfill(4)
for i in range(len(df['ZIP'])):
    df['ZIP'][i] = df['ZIP'][i].zfill(5)
```

```python
df.head()
```

```python
df.loc[df['ZIP'] == '00603', 'total_pop']
```

```python
conn = sqlite3.connect('data/zips.db')

# write the data to a new table in the database
df.to_sql('factors', conn, if_exists='replace', index=False)
conn.close()
```
